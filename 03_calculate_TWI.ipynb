```python
import numpy as np
import elevation
import richdem as rd
from osgeo import gdal, gdalconst
import rioxarray
#https://www.earthdatascience.org/tutorials/get-slope-aspect-from-digital-elevation-model/
#https://richdem.readthedocs.io/en/latest/flow_accumulation.html
```

```python
#dem_path = 'C:/Uni Jena/GEO 411/Data/DGM/Thueringen/2010_2013/merged_result.tif'
#output_file = 'C:/Uni Jena/GEO 411/Data/DGM/Thueringen/2010_2013/merged_result.map'
```

```python
def TWI(path, infile, outfile):
    #load DEM
    dem = rd.LoadGDAL(path+'/'+infile,no_data=0.0)
    #calculate slope
    slope = rd.TerrainAttribute(dem, attrib='slope_radians')
    #Fill depressions with epsilon gradient to ensure drainage
    rd.FillDepressions(dem, epsilon=True, in_place=True)
    #Get flow accumulation with no explicit weighting. The default will be 1.
    accum_d8 = rd.FlowAccumulation(dem, method='D8')
    #get DEM as xarray
    xds = rioxarray.open_rasterio(path+'/'+infile)
    #get pixelsize from xarray DEM
    cellar = float(xds.x[1]-xds.x[0])
    # calculate twi and save to file
    twi = np.log((accum_d8 * cellar) / np.tan(slope))
    #eliminate inf values/ now there are gaps in raster
    twi[twi==np.inf] = np.nan
    #write TWI to file
    rd.SaveGDAL(path+'/'+outfile, twi)
    dem,slope,accum_d8,xds,cellar,twi = None, None, None, None, None, None
```
