```python
from rasterio.plot import show
from rasterio.merge import merge
import rasterio as rio
from rasterio import features
import rioxarray

from pathlib import Path
import glob
import xarray as xr
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import geopandas as gpd
```

```python
def make_radolan_xarray(path, str_start, str_end, mini, maxi):
    #path = 'C:/Uni Jena/GEO 411/Data/Radolan/Thueringen/'
    #getall file paths from path
    file_list = glob.glob(path + "raa01*.tif")

    # make empty list
    array_list = []
    
    for i in file_list[mini:maxi]:
        #open ds from filepath
        #da = xr.open_rasterio(i)
        da = rioxarray.open_rasterio(i)
        #convert date string from filename to timestamp
        time = pd.to_datetime(i[str_start:str_end], format='%y%m%d%H%M')
        # assign timestamp value to variable time
        da = da.assign_coords({"time": (time)})
        #da = da.expand_dims(dim="time")
        #delete default variable band
        #da = da.drop_vars("band")
        #put array in list
        array_list.append(da)

    #concatenate arrays from list to one xarray    
    radolan_xr = xr.concat([array_list[x] for x in range(len(array_list))], dim = "time")
    return radolan_xr
```

```python
clip0_list = []
clip1_list = []
clip2_list = []
clip3_list = []
clip4_list = []
clip5_list = []
clip6_list = []
average = []

#get timestamps
time_list_thuer = list(clipped0.time.values)

#for each dataset get mean values for each scene
for i in range(len(clipped0)):
    a0 = float(clipped0[i].mean())
    a1 = float(clipped1[i].mean())
    a2 = float(clipped2[i].mean())
    a3 = float(clipped3[i].mean())
    a4 = float(clipped4[i].mean())
    a5 = float(clipped5[i].mean())
    a6 = float(clipped6[i].mean())
    a = float(radolan_thueringen_new[i].mean())
    
    clip0_list.append(a0)
    clip1_list.append(a1)
    clip2_list.append(a2)
    clip3_list.append(a3)
    clip4_list.append(a4)
    clip5_list.append(a5)
    clip6_list.append(a6)
    average.append(a)
```
